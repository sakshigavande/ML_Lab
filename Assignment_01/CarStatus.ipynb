```python
fuel_type = ['CNG','Petrol','Diesel','LPG','Electric']
cars = [62,3325,3852,12,2]
colors = ['yellow','pink','Black','orange','white']
mt.pie(cars,labels=fuel_type,colors=colors)
mt.title('Car by fuel_type')
mt.axis('equal')
mt.show()
```

```python
cars = [62,3325,3852,12,2]
mt.hist(cars, bins=len(cars), color='pink', edgecolor='black')
mt.xlabel('Number of cars')
mt.ylabel('Frequency')
mt.title('Histogram of Cars by Fuel Type')
mt.savefig('C:/Users/SAI/Desktop/ML.png')  
mt.show()
```

```python
yearly_total_km = data.groupby('Year')['Kilometers_Driven'].sum()
print(yearly_total_km)
```

```python
mt.hist(data['Year'], bins=range(data['Year'].min(), data['Year'].max()+1 ), color='skyblue', edgecolor='blue')
mt.xlabel('Year')
mt.ylabel('Number of Cars')
mt.title('Number of Cars by Year')
mt.show()
```

```python
yearly_km = data.groupby('Year')['Kilometers_Driven'].sum().sort_index()
mt.bar(yearly_km.index, yearly_km.values, color='lightgreen', edgecolor='black')
mt.xlabel('Year')
mt.ylabel('Total Kilometers Driven')
mt.title('Year-wise Total Kilometers Driven')

mt.show()
```

```python
yearly_price = data.groupby('Year')['Price'].mean().sort_index()
mt.plot(yearly_price.index, yearly_price.values, marker='o', linestyle='-', color='purple')
mt.xlabel('Year')
mt.ylabel('Average Car Price (in lakhs)')
mt.title('Year-wise Average Car Price')
mt.grid(True)
mt.show()
```

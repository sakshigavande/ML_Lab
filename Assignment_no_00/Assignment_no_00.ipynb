```python
choice = int(input("Select the operation you want to perfrom:-\n1.Addition\n2.Subtraction\n3.Multiplication\n4.Division\n"));
a=int(input("Enter the value of A:- "));
b=int(input("Enter the value of b:- "));
match choice: 
 case 1:
        c=a+b;
        print(c);
 case 2:
        c=a-b;
        print(c);
 case 3:
        c=a*b;
        print(c);
 case 4:
        c=a/b;
        print(c);
 case _:
        print("Invalid Input");
```
